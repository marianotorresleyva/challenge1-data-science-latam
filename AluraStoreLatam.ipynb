### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

#1. Análisis de facturación



```python
# Suma Total de Precio
total_tienda = tienda['Precio'].sum()
print(total_tienda)

#Productos vendidos por categoria
total_producto_categoria = tienda['Categoría del Producto'].value_counts()
print(total_producto_categoria)

#Valoracion Media Por tienda
valoracion_tienda = tienda['Calificación'].mean()
print(valoracion_tienda)

#Producto más vendido de la tienda
producto_mas_vendido = tienda['Producto'].value_counts().idxmax()
print(producto_mas_vendido)

#Producto menos vendido de la tienda
producto_menos_vendido = tienda['Producto'].value_counts().idxmin()
print(producto_menos_vendido)

#Valor Promedio de Envio Por Tienda
valor_promedio_envio = tienda['Costo de envío'].mean()
print(valor_promedio_envio)
```

```python
#Grafica comparando el producto más vendido con el menos vendido

data_tienda = {'Producto': [producto_mas_vendido, producto_menos_vendido],
                'Cantidad': [tienda['Producto'].value_counts()[producto_mas_vendido],
                         tienda['Producto'].value_counts()[producto_menos_vendido]]}
df_tienda = pd.DataFrame(data_tienda)
df_tienda.plot(kind='barh', x='Producto', y='Cantidad')
plt.show()
```
