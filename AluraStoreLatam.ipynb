# 2. Ventas por categoría

```python
print("\nTienda 1:")
total_producto_categoria1 = tienda1['Categoría del Producto'].value_counts()
print(total_producto_categoria1)

print("\nTienda 2:")
total_producto_categoria2 = tienda2['Categoría del Producto'].value_counts()
print(total_producto_categoria2)

print("\nTienda 3:")
total_producto_categoria3 = tienda3['Categoría del Producto'].value_counts()
print(total_producto_categoria3)

print("\nTienda 4:")
total_producto_categoria4 = tienda4['Categoría del Producto'].value_counts()
print(total_producto_categoria4)
```

# 3. Calificación promedio de la tienda


```python
valoracion_tienda1 = tienda1['Calificación'].mean()
print(f"Calificación Promedio Tienda 1: {valoracion_tienda1:.2f}")

valoracion_tienda2 = tienda2['Calificación'].mean()
print(f"Calificación Promedio Tienda 2: {valoracion_tienda2:.2f}")

valoracion_tienda3 = tienda3['Calificación'].mean()
print(f"Calificación Promedio Tienda 3: {valoracion_tienda3:.2f}")

valoracion_tienda4 = tienda4['Calificación'].mean()
print(f"Calificación Promedio Tienda 4: {valoracion_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
print("\nTienda 1:")
conteo_productos1 = tienda1['Producto'].value_counts()
producto_mas_vendido1 = conteo_productos1.idxmax()
cant_mas_vendido1 = conteo_productos1.max()
print(f"  Más vendido: {producto_mas_vendido1} (Cantidad: {cant_mas_vendido1})")

producto_menos_vendido1 = conteo_productos1.idxmin()
cant_menos_vendido1 = conteo_productos1.min()
print(f"  Menos vendido: {producto_menos_vendido1} (Cantidad: {cant_menos_vendido1})")
```

```python
print("\nTienda 2:")
conteo_productos2 = tienda2['Producto'].value_counts()
producto_mas_vendido2 = conteo_productos2.idxmax()
cant_mas_vendido2 = conteo_productos2.max()
print(f"  Más vendido: {producto_mas_vendido2} (Cantidad: {cant_mas_vendido2})")

producto_menos_vendido2 = conteo_productos2.idxmin()
cant_menos_vendido2 = conteo_productos2.min()
print(f"  Menos vendido: {producto_menos_vendido2} (Cantidad: {cant_menos_vendido2})")
```

```python
print("\nTienda 3:")
conteo_productos3 = tienda3['Producto'].value_counts()
producto_mas_vendido3 = conteo_productos3.idxmax()
cant_mas_vendido3 = conteo_productos3.max()
print(f"  Más vendido: {producto_mas_vendido3} (Cantidad: {cant_mas_vendido3})")

producto_menos_vendido3 = conteo_productos3.idxmin()
cant_menos_vendido3 = conteo_productos3.min()
print(f"  Menos vendido: {producto_menos_vendido3} (Cantidad: {cant_menos_vendido3})")
```
